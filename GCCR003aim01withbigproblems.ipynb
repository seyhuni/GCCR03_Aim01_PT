# DATE DELTA

```python
data['Date_of_onset'] = pd.to_datetime(data['Date_of_onset'])
```

```python
new_data=data.dropna(subset = ['Smell_current', 'Smell_before_illness', 'Smell_current_Time_Stamp', 'Date_of_onset'])
new_data.describe()
```

```python
data['days_from_onset'] = ((data['Smell_current_Time_Stamp'] - data['Date_of_onset']).dt.total_seconds()/(60*60*24))
data['days_from_onset'].head(8)
data['days_from_onset'].describe()
```

```python
data["Date_of_onset"] = pd.to_datetime(data["Date_of_onset"], dayfirst=True)
X = data[['days_from_onset']] #time duration between before and follow up
y = data['Smell_before_illness'] - data['Smell_current']

y.describe()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
import matplotlib.pyplot as plt
print('Coefficients: \n', lm.coef_)
predictions = lm.predict(X_test)
plt.scatter(y_test, predictions)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
plt.show()
plt.plot(y)
plt.show()
plt.plot(predictions)
plt.show()
```

```python
from sklearn import metrics
import numpy as np

print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
long_h = data.query('percentage_recovery_smell < 80').reset_index(drop=True)
status_lh = pd.DataFrame(np.zeros((len(long_h), )), columns=['status'])
lh_df = pd.concat([long_h, status_lh], axis=1)

normal = data.query('percentage_recovery_smell >= 80').reset_index(drop=True)
status_n = pd.DataFrame(np.ones((len(normal), )), columns=['status'])
n_df = pd.concat([normal, status_n], axis=1)

final_df = pd.concat([lh_df, n_df], axis=0)

final_df.describe()
```

```python
import matplotlib.pyplot as plt
from sksurv.nonparametric import kaplan_meier_estimator
# 0 : long, 1 : normal
final_df['status'] = final_df['status'].astype('bool')
time, survival_prob = kaplan_meier_estimator(final_df["status"], final_df["days_from_onset"])
plt.step(time, survival_prob, where="post")
plt.ylabel("est. probability of survival $\hat{S}(t)$")
plt.xlabel("time $t$")
plt.xlim((-332,500))
```
